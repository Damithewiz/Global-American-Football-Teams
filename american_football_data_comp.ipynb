## Data Validation & Cleaning

```python
# Standardize formatting (title case for names, strip whitespace)
cols_to_title = ["Team Name", "City", "Home Field/Stadium", "Head Coach"]
for col in cols_to_title:
    df[col] = df[col].astype(str).str.strip().str.title()

# Ensure League and Conference are also clean
df["League"] = df["League"].str.strip()
df["Conference"] = df["Conference"].str.strip()

# Check for nulls in critical fields
critical_columns = ["Team Name", "City", "Country", "Home Field/Stadium", "Head Coach"]
null_report = df[critical_columns].isnull().sum()
print("Nulls in critical fields:\n", null_report)

# Drop or flag duplicates based on team and city
duplicates = df.duplicated(subset=["Team Name", "City"])
print(f"Duplicate rows found: {duplicates.sum()}")

# Standardize country names
country_corrections = {
    "Usa": "United States Of America",
    "United States": "United States Of America",
    "UK": "United Kingdom",
    "England": "United Kingdom",
    "Deutschland": "Germany"
}
df["Country"] = df["Country"].replace(country_corrections).str.strip().str.title()

# Sanity check: Unique values in each column
print("Unique leagues:", df["League"].unique())
print("Unique countries:", df["Country"].unique())
```

```python
# Identify duplicates based on "Team Name" and "City"
duplicates = df[df.duplicated(subset=["Team Name", "City"], keep=False)]

# Display the full duplicated rows
print("🔁 Duplicated Rows:")
print(duplicates)
```

```python
# Save the cleaned dataset
df.to_csv("cleaned_football_teams.csv", index=False)

print(" Dataset cleaned and saved successfully!")
```

## Geocode Locations for Mapping

```python
# !pip install geopy
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="football_mapper")

def geocode(city_country):
    try:
        location = geolocator.geocode(city_country)
        return pd.Series([location.latitude, location.longitude])
    except:
        return pd.Series([None, None])

df[['Latitude', 'Longitude']] = df.apply(
    lambda row: geocode(f"{row['City']}, {row['Country']}"), axis=1
)
```

```python
print(df)
```

```python
failed_geocodes = df[df['Latitude'].isnull() | df['Longitude'].isnull()]
print("Rows with missing coordinates:")
print(failed_geocodes[['Team Name', 'City', 'Country', 'Latitude', 'Longitude']])
```

```python
print(failed_geocodes.count())
```

```python
from geopy.exc import GeocoderTimedOut, GeocoderServiceError
import time

# Initialize geolocator
geolocator = Nominatim(user_agent="geo_inline_update")

# Loop through rows with missing coordinates
for idx, row in failed_geocodes.iterrows():
    city = row['City']
    country = row['Country']
    team = row['Team Name']
    
    print(f" Trying: {city}, {country}")
    
    try:
        result = geolocator.geocode(f"{city}, {country}", timeout=10)
        if result:
            print(f" {team}: {result.latitude}, {result.longitude}")
            # Update the main DataFrame (not just failed_geocodes)
            df.loc[idx, 'Latitude'] = result.latitude
            df.loc[idx, 'Longitude'] = result.longitude
        else:
            print(f" {team}: Not found")
    except (GeocoderTimedOut, GeocoderUnavailable):
        print(f" Timeout or Service Error on: {team}")
    
    # Respect rate limit
    time.sleep(2)
```
